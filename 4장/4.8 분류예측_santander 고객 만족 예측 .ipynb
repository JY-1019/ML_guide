```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
import matplotlib
import matplotlib.pyplot as plt
import warnings
%matplotlib inline
warnings.filterwarnings('ignore')

cus_df = pd.read_csv('./santander-customer-satisfaction/train_santander.csv', 
                  encoding = 'latin-1')
```

```python
cus_df.head()
```

```python
cus_df.describe()
```

```python
cus_target = cus_df['TARGET']
cus_features = cus_df.drop('TARGET', axis = 1)
print(cus_target.head())
cus_features.head()
```

```python
cus_features['var3'].value_counts()
```

```python
cus_features['var3'].replace(-999999, 2, inplace=True)
cus_features['var3'].value_counts()
```

```python
X_train, X_test, y_train, y_test = train_test_split(cus_features, cus_target,
                                                   test_size = 0.2, random_state=156,
                                                   stratify=cus_target)
# eval_set 에서 쓰기 위해 데이터를 짤라 뒀으나, 데이터량이 부족해서 문제가 있음.
evals_count = int(np.round(y_train.count()*0.1))
X_train_evals=X_train.iloc[:evals_count,:]
X_train_real=X_train.iloc[evals_count:,:]
y_train_evals=y_train.iloc[:evals_count]
y_train_real=y_train.iloc[evals_count:]
```

# XGB

```python
from xgboost import XGBClassifier
from sklearn.metrics import roc_auc_score

evals = [(X_train_real, y_train_real),(X_train_evals, y_train_evals)]

XGB_model = XGBClassifier(n_estimators = 500, random_state = 156)
XGB_model.fit(X_train, y_train, early_stopping_rounds=100, eval_metric='auc',
             eval_set = evals)
xgb_roc_score = roc_auc_score(y_test, XGB_model.predict_proba(X_test)[:,1],
                             average = 'macro')
print('ROC AUC : {0:.4f}'.format(xgb_roc_score))
```

```python
from sklearn.model_selection import GridSearchCV

XGB_model = XGBClassifier(n_estimators = 100, random_state = 156)

params = {
    'max_depth' : [5, 7],
    'min_child_weight' : [1, 3],
    'colsample_bytree' : [0.5, 0.75]
}

# 수행 속도 고려해서 cv를 지정하지않음.
grid_model=GridSearchCV(XGB_model,param_grid = params)
grid_model.fit(X_train, y_train, early_stopping_rounds = 30, eval_set = evals,
              eval_metric = 'auc')

print('GridSearchCV 최적 파라미터', grid_model.best_params_)

xgb_roc_score = roc_auc_score(y_test, grid_model.predict_proba(X_test)[:,1],
                           average = 'macro')
print('ROX AUC : {0:.4f}'.format(xgb_roc_score))
```

```python
# 해당 결과 값을 n_estimators 증가, learning_rate = 0.02, reg_alpha = 0.03
xgb_best_model = XGBClassifier(n_estimators = 1000, random_state = 156,
                              learning_rate = 0.02, max_depth=5, min_child_weight=1,
                              colsample_bytree = 0.75, reg_alpha = 0.03)

# 연산 시간 줄이기 위해서 early stopping 20 으로 설정
xgb_best_model.fit(X_train, y_train, early_stopping_rounds = 20, eval_set = evals,
                  eval_metric = 'auc')
```

```python
xgb_roc_score = roc_auc_score(y_test, xgb_best_model.predict_proba(X_test)[:,1],
                              average = 'macro')
print('ROC AUC Score : {0:.4f}'.format(xgb_roc_score))
```

```python
from xgboost import plot_importance
import matplotlib.pyplot as plt
%matplotlib inline

fig, ax = plt.subplots(1, 1, figsize = (10, 8))
plot_importance(xgb_best_model, ax = ax, max_num_features = 20, height = 0.4)
```

# LightGBM

```python
from sklearn.metrics import confusion_matrix, accuracy_score
from sklearn.metrics import recall_score, precision_score
from sklearn.metrics import f1_score, roc_auc_score

def get_clf_eval(y_test, pred=None, pred_proba=None):
    confusion = confusion_matrix(y_test, pred)
    accuracy = accuracy_score(y_test, pred)
    precision = precision_score(y_test, pred)
    recall = recall_score(y_test, pred)
    f1 = f1_score(y_test, pred)
    roc_auc = roc_auc_score(y_test, pred_proba)
    print('오차행렬')
    print(confusion)
    print('정확도: {0:.4f}, 정밀도: {1:.4f}, 재현율: {2:.4f},\
    F1: {3:.4f}, AUC:{4:.4f}'.format(accuracy, precision, recall, f1, roc_auc))
```

```python
from lightgbm import LGBMClassifier

lgbm_model = LGBMClassifier(n_estimators = 200 )

lgbm_model.fit(X_train, y_train, early_stopping_rounds=100, eval_metric = 'auc',
              eval_set = evals, verbose = True)
```

```python
lgbm_roc_score = roc_auc_score(y_test, lgbm_model.predict_proba(X_test)[:,1],
                              average = 'macro')
print('roc auc : {0:.4f}'.format(lgbm_roc_score))
```

```python
from sklearn.model_selection import GridSearchCV

params = {
    'num_leaves' : [32, 64],
    'max_depth' : [128, 160],
    'min_child_samples' : [60, 100],
    'subsample' : [0.8, 1]
}
grid_lgbm_model = GridSearchCV(lgbm_model, param_grid = params)
grid_lgbm_model.fit(X_train, y_train, early_stopping_rounds = 30, eval_metric='auc',
                   eval_set = evals)
```

```python
print('Best parameters : ',grid_lgbm_model.best_params_)
lgbm_roc_score = roc_auc_score(y_test, grid_lgbm_model.predict_proba(X_test)[:,1],
                              average = 'macro')
print('ROC AUC Score:', lgbm_roc_score)
```

```python
lgbm_best_model = LGBMClassifier(n_estimators = 500, 
                                 max_depth = 128,
                                 min_child_samples = 100,
                                 num_leaves = 64,
                                 subsample = 0.8)

lgbm_best_model.fit(X_train, y_train, early_stopping_rounds=100, eval_metric='auc',
                   eval_set = evals, verbose = True)
lgbm_roc_best_score = roc_auc_score(y_test, lgbm_best_model.predict_proba(X_test)[:,1],
                                   average = 'macro')
print('ROC AUC SCORE : ', lgbm_roc_best_score)
# best score 가 앞에서보다 오히려 떨어졌는데 n_estimator 값이 너무 작기 때문이라고 추정된다/
```
